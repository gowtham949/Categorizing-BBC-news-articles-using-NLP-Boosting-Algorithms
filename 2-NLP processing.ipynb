```python
news_Art='Summarized_news_Art.csv'
import pandas as news_ArtP

news_Art_F = news_ArtP.read_csv(news_Art)
news_Art_F
```

```python
news_Art_F= news_Art_F.dropna()
news_Art_F.shape
```

#### NLP process

```python
import nltk as news_Art_Fnlt
news_Art_Fnlt.download('stopwords')
import regex as news_Art_Frge
from nltk.corpus import stopwords as news_Art_Fspwd
from nltk.stem.porter import PorterStemmer as news_Art_FPst
```

```python
news_Art_Fspwd_W = news_Art_Fspwd.words('english')
news_Art_FPst_W = news_Art_FPst() # Stemming variable
news_At=[]
```

```python
def bbc_n(uj):
    uj = uj.lower()
    uj = news_Art_Frge.sub("[^a-z]+", " ", uj)
    uj = " ".join([u for u in uj.split() if u not in news_Art_Fspwd_W])
    uj=' '.join([news_Art_FPst_W.stem(u) for u in uj.split()])
    return uj
```

```python

for s in news_Art_F['News']:
  news_At.append(bbc_n(s))
news_Art_F['News']=news_At
news_Art_F.head(n=4)
```
